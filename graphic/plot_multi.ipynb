# プロット複数表示

```python
# ライブラリのロード
import numpy as np
import matplotlib.pyplot as plt
```

```python
#　軸の作成
x1 = [i for i in range(100)]
y1 = [i for i in range(100)]
x2 = [i for i in range(100)]
y2 = [i * (-1) for i in range(100)]
```

```python
# グラフ領域を準備
fig = plt.figure(figsize = (15, 15))

# 描画領域の作成
# 行数, 列数, 位置
ax1 = fig.add_subplot(2, 1, 1)
ax2 = fig.add_subplot(2, 1, 2)

# ラベル
label1 = "plot1"
label2 = "plot2"

# 描画する
ax1.plot(x1, y1, label = label1)
ax2.plot(x2, y2, label = label2)

# xラベルを追加
ax1.set_xlabel(label1, fontsize=15)
ax2.set_xlabel(label2)

# yラベルを追加
ax1.set_ylabel(label1)
ax2.set_ylabel(label2)

# タイトルの追加
ax1.set_title("sample1", fontsize=15)
ax2.set_title("sample2", fontsize=20)

# ticksの調整
# https://qiita.com/tsukada_cs/items/8d31a25cd7c860690270
ax2.set_xticklabels(["", "A", "B", "C", "D", "E", "F"])
ax2.set_yticklabels("")
ax2.tick_params(left=False)

# レジェンドの位置
ax1.legend(loc = 'upper right')
ax2.legend(loc = 'lower left')

plt.show()
```

## 応用
### n x m のプロット

```python
# 行数、列数
N = 5
M = 4

# グラフの描画領域
fig = plt.figure(figsize = (15,15))

# カラーマップインスタンス生成
cm = plt.get_cmap("hsv")
# シードの固定
np.random.seed(seed=32)

for i in range(1, N+1):
    for j in range(1, M+1):
        # 添え字に注意
        ax = fig.add_subplot(N, M, N * (j - 1) + i)
        col = cm((N * (j - 1) + i) / (N * M))
        x = np.random.rand(100)
        y = np.random.rand(100)
        ax.scatter(x, y, color=col)
```
